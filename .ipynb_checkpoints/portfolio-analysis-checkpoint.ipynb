# Portfolio Risk and Return Analysis

https://codingandfun.com/portfolio-risk-and-returns-python/

```python
#Daily percentage change
return_assets = prices.pct_change()
return_assets.head()
```

```python
#Mean average returns
daily_returns_portfolio_mean = return_assets.mean()
print(daily_returns_portfolio_mean)
```

```python
#Initial portfolio weights
initial_weight = np.array([0.1201,0.254,0.122,0.072,0.410,0.007,0.004,0.007])
```

```python
#Calculate average daily portfolio return given initial allocation
allocated_daily_returns = (initial_weight * daily_returns_portfolio_mean)

portfolio_return = np.sum(allocated_daily_returns)
print("Average Daily Portfolio Return: {} %".format(round(portfolio_return*100,2)))
```

```python
# calculate portfolio daily returns
return_assets['portfolio_daily_returns'] = return_assets.dot(initial_weight)
return_assets
```

```python
Cumulative_returns_daily = (1+return_assets).cumprod()
Cumulative_returns_daily.tail(5)
```

```python
#portfolio cummulative returns plot
Cumulative_returns_daily['portfolio_daily_returns'].plot()
```

```python
#calculate the daily covariance and then we annualise it by multiplying for 252 (i.e. 252 are more or less the trading days in a year)
matrix_covariance_portfolio = return_assets.iloc[:,:-1]
matrix_covariance_portfolio = (matrix_covariance_portfolio.cov())*252

matrix_covariance_portfolio
```

```python
portfolio_variance = np.dot(initial_weight.T,np.dot(matrix_covariance_portfolio, initial_weight))

#standard deviation (risk of portfolio)
portfolio_risk = np.sqrt(portfolio_variance)
portfolio_risk
```

#### It would be interesting to check if we could achieve a similar return lowering the risk.

# Portfolio Optimization

The higher the Sharpe Ratio, the better a portfolio returns have been relative to the taken risk

https://codingandfun.com/portfolio-optimization-with-python/

```python
matrix_covariance_portfolio.keys()
```

```python
number_of_portfolios = 2000
RF = 0 #risk free return 

portfolio_returns = []
portfolio_risk = []
sharpe_ratio_port = []
portfolio_weights = []

for portfolio in range (number_of_portfolios):
    weights = np.random.random_sample((len(matrix_covariance_portfolio.keys())))
    #print(type(weights))
    #print(weights)
    #below line ensures that the sum of our weights is 1
    weights = weights / np.sum(weights)
    annualize_return = np.sum((return_assets.iloc[:,:-1].mean() * weights) * 252)
    
    portfolio_returns.append(annualize_return)
    #variance
    matrix_covariance_portfolio = (return_assets.iloc[:,:-1].cov())*252
    portfolio_variance = np.dot(weights.T,np.dot(matrix_covariance_portfolio, weights))
    portfolio_standard_deviation= np.sqrt(portfolio_variance) 
    portfolio_risk.append(portfolio_standard_deviation)
    #sharpe_ratio
    sharpe_ratio = ((annualize_return- RF)/portfolio_standard_deviation)
    sharpe_ratio_port.append(sharpe_ratio)

    #keep weights as well to find out later the weights from the optimized portfolio
    portfolio_weights.append(weights)
```

```python
#convert to arrays
portfolio_risk = np.array(portfolio_risk)
portfolio_returns = np.array(portfolio_returns)
sharpe_ratio_port = np.array(sharpe_ratio_port)
```

```python
#Plotting portfolios
plt.figure(figsize=(10, 5))
plt.scatter(portfolio_risk, portfolio_returns, c=portfolio_returns / portfolio_risk) 
plt.xlabel('volatility')
plt.ylabel('returns')
plt.colorbar(label='Sharpe ratio')
```

**Sharpe Ratio â€“ Random Portfolios**

In the graph, each point represents a portfolio. We see that portfolios with the higher Sharpe Ratio are shown as yellow. Yellow coloured portfolios are preferable since they offer better risk adjusted returns.

```python
porfolio_metrics = [portfolio_returns,portfolio_risk,sharpe_ratio_port, portfolio_weights] 
#from Python list we create a Pandas DataFrame
portfolio_dfs = pd.DataFrame(porfolio_metrics)
portfolio_dfs = portfolio_dfs.T
#Rename the columns:
portfolio_dfs.columns = ['Port Returns','Port Risk','Sharpe Ratio','Portfolio Weights']

#convert from object to float the first three columns.
for col in ['Port Returns', 'Port Risk', 'Sharpe Ratio']:
    portfolio_dfs[col] = portfolio_dfs[col].astype(float)
portfolio_dfs
```

```python
#portfolio with the highest Sharpe Ratio
Highest_sharpe_port = portfolio_dfs.iloc[portfolio_dfs['Sharpe Ratio'].idxmax()]
#portfolio with the minimum risk 
min_risk = portfolio_dfs.iloc[portfolio_dfs['Port Risk'].idxmin()]

print(Highest_sharpe_port)
print("")
print(min_risk)
```

```python
#Allocation that maximizes Sharpe Ratio
Highest_sharpe_port["Portfolio Weights"]
```

```python
#Summarizing Optimal Action to take
action = pd.Series(data=np.subtract(Highest_sharpe_port["Portfolio Weights"],initial_weight)*100, index=prices.keys())
action
```
